```python
#Reading data from a csv file
import pandas as readFile
foot_ball = readFile.read_csv("C:\\Users\\Luzuko\\Downloads\\dataset - 2020-09-24.csv")
print(foot_ball.head(5))
```

```python
#identifying missing values in columns
missing_values = foot_ball.isnull().sum()
print(missing_values)
```

```python
#finding duplicate values and count duplicate rows
duplicate_values = foot_ball.duplicated().sum()
print(f"Number of duplicate entries {duplicate_values}")
```

```python
#Creating a Scatter plot of goals and assists
import matplotlib.pyplot as plt

#plt.figure(figsize=(20, 50))
plt.scatter(foot_ball['Goals'], foot_ball['Assists'])
plt.title('Scatter Plot of Goals VS Assists')
plt.ylabel('Goals')
plt.xlabel('Assists')
plt.show()
```

```python
Creating a histogram of goals
```

```python
plt.hist(foot_ball['Goals'], bins = 10, color = 'blue', alpha = 0.5)
plt.title('Histogram of Goals')
plt.ylabel('Freqency')
plt.xlabel('Goals')
plt.show()
```

```python
Creating a bar chart of names and age
```

```python
plt.figure(figsize = (20,5))
plt.bar(foot_ball['Name'], foot_ball['Age'], color = 'blue', alpha = 0.7)
plt.title('Names of players by their age')
plt.xlabel('Names')
plt.ylabel('Ages')
plt.show()
```

```python
#Creating pie chart

#getting a sample data of all teams found in the club column
teams = foot_ball['Club'].unique()
tackles = foot_ball['Shooting accuracy %'].unique()



print(teams)
print(tackles)
```
